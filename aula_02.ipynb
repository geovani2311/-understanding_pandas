## Loc e Iloc

```python
import pandas as pd
import numpy as np
from numpy.random import randn
```

```python
# Gerando tabela
df = pd.DataFrame(randn(5,4), index = ["A","B","C","D","E"], columns = "W X Y Z".split())
```

```python
df
```

```python
df.loc[['A','B'],['W']]
```

```python
#selecionando DataFrame com como se fosse pegar valores de uma lista:
df.iloc[0,2]
# sendo no caso acima, 0 a linha e 2 a coluna, fazendo com que pegue valor de A,Y
```

```python
#sabendo disso temos diversas maneiras de coletar dados especificos dentro de um dataframe
df.iloc[1:2, 1:3]
```

**seleções condicionais**

```python
df > 0
```

```python
df [df>0]
```

```python
df[df ['Y'] > 0]
```

```python
df [(df ['Y'] > 0) & (df ['Z'] > 0)]
```
